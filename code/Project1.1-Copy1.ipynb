```python
%matplotlib qt5
#%matplotlib notebook
from modsim import *
from pandas import read_html
import pandas
import numpy
```

```python
filename = 'data/returnedsalmonpopulation.html'
tables = read_html(filename, header=1, index_col=0, decimal='M')
```

```python
table1 = tables[0]
table1
```

```python
total = table1.Total
```

```python
Time = System(t0 = 1967, t_end = 2014)
habitat_destruction = 0.5
total = table1.Total
```

```python
population_change = TimeSeries() 
population_change = total.diff()
population_change[Time.t0] = 100
population_change.columns = {' Year', 'dp/dt'}
population_change
```

```python
def general_growth(pop):
    pop += pop*1.5
    return pop
```

```python
def update_func(pop, t):
    net_growth2 = population_change.sample(n=1)
    pop += net_growth2
    if pop.values < 0:
        pop.values = 0*pop.values
    return pop.values
```

```python
def run_simulation(pop, t, p):
    pop1 = general_growth(pop)
    if flip(p):
        pop2 = update_func(pop1, t)
    else: 
        pop2 = pop1
    return pop2
run_simulation(total, 0, 0.5)
```

```python
def run_model(p):
    salmon_population = System(pop0 = 781)
    model = TimeSeries()
    model[Time.t0] = salmon_population.pop0
    for t in linrange(Time.t0, Time.t_end):
        model[t+1] = run_simulation(model[t], t, p)
    Time.model = model
```

```python
run_model(0.5)
print(Time.model)
plot(total, '-', color='darkblue', label='Raw data')
plot(Time.model, '-', color='red', label='model')
def decorate():
    """Add a legend and label the axes.
    """
    legend(loc='best')
    label_axes(title='Returned Salmon Population Over Time',
               xlabel='Year', 
               ylabel='Salmon Population')
decorate()
```
